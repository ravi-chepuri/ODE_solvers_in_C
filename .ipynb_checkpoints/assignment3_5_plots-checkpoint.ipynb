```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/projectile.dat', index_col=0)
```

```python
df.head()
```

```python
plt.figure(figsize=(6.4*2, 4.8*2))

colors = ['b', 'g', 'r', 'c', 'm', 'y']

i=0
for angle in [30, 35, 40, 45, 50, 55]:
    xcol = "{} deg x".format(angle)
    ycol = "{} deg y".format(angle)
    lab = "${}^\\circ$".format(angle)
    plt.plot(df[xcol], df[ycol], ls='none', marker = '.', color=colors[i], label = lab)
    i += 1
    
g = -9.8
xs = []
ys = []
lab = []
ts = np.linspace(0, 150, 150)
i = 0
for angle in [30, 35, 40, 45, 50, 55]:
    vx = 700 * np.cos(angle * np.pi/180)
    vy = 700 * np.sin(angle * np.pi/180)
    xs.append(ts * vx)
    ys.append(ts * vy + 1/2 * g * np.square(ts))
    
    lab.append("${}^\\circ$ (analytical)".format(angle))

for i in range(6):
    plt.plot(xs[i], ys[i], color=colors[i], label = lab[i])
    
    

plt.xlim(0, 60000)
plt.ylim(0, 20000)
plt.title("Euler's method for projectile motion (no air resistance)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(title="Launch angle")

plt.show()
```

Above, the Euler's method solution to projectile motion is plotted, along with the analytical solution. A number of initial launch angles are used, each with an initial velocity of 700m/s. Note that Euler's method consistently overestimates the height of the projectile -- this is an inherent limitation of the method. Usage of a method such as order 4 Runge Kutta could help solve this issue. Compare to Fig. 2.4 a) in Computational Physics.
