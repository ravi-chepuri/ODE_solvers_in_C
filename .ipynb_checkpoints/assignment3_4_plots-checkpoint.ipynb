```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tauB_small = pd.read_csv('data/twospecies_euler_tauBsmall.dat', index_col=0)
tauB_large = pd.read_csv('data/twospecies_euler_tauBlarge.dat', index_col=0)
```

```python
tauB_small.plot(ls='none', marker = '.')
plt.xlabel('Time (sec)')
plt.ylabel('Nuclei remaining')
plt.title('Two-species decay with $\\tau_B << \\tau_A$')
```

Qualitative explanation: This plot shows the outcome of a two-species radioactive decay process where species A decays to species B over a comparitively long time, and species B decays away over a comparitively short time. At the beginning, we see that the number of B nuclei quickly drops to near zero, as we expect. Then, as the number of A nuclei slowly lowers, those that have decayed into B quickly decay away. 

```python
tauB_large.plot(ls='none', marker = '.')
plt.xlabel('Time (sec)')
plt.ylabel('Nuclei remaining')
plt.title('Two-species decay with $\\tau_B >> \\tau_A$')
```

Qualitative explanation: This plot shows the outcome of a two-species radioactive decay process where species A decays to species B over a comparitively short time, and species B decays away over a comparitively long time. At the beginning, we see that the number of A nuclei quickly decreases to near zero, as we expect. These A nuclei are converted into B nuclei, causing a spike in the number of B nuclei, as we observe. Then after some time, we see that these B nuclei decay over a longer time scale.
