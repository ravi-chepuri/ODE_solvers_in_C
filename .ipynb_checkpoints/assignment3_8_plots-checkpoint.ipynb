```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
qList = [0.1, 0.2, 0.5, 1.0]
omegaDList = [0.25, 0.5, 0.75, 0.8, 0.85, 0.9, 0.92, 0.95, 0.96, 0.97, 0.98, 0.99, 1.0, 1.01, 1.02, 1.03, 1.04, 1.05, 1.08, 1.1, 1.15, 1.2, 1.25, 1.5, 1.75]
```

```python
data = {}

for q in qList:
    data[q] = {}
    for omega_d in omegaDList:
        filename = f"./data/assignment3_8/{q}_{omega_d}.dat"
        data[q][omega_d] = pd.read_csv(filename)
```

```python
plt.plot(data[0.5][1]["t"], data[0.5][1]["theta_rk4"], ls='none', marker='.')

plt.title("A damped driven harmonic oscillator")
plt.xlabel("$t$")
plt.ylabel(r"$\theta$")
plt.xlim(0, 25)
```

This plot shows the motion of damped driven harmonic oscillator described by the equation
$$ \frac{d^2\theta}{dt^2} = - \frac{g}{l} \theta - q \frac{d\theta}{dt} + F_D \sin{\Omega_D t}$$


as solved via a Order 4 Runge Kutta method. Note that the motion eventually stabilizes to a sinusoidal steady state solution. In the next plot we analyze the dependence of the steady state amplitude on the parameters $q$ and $\Omega_D$.

```python
def steady_state_amplitude(dataframe):
    """Return a measurement of the steady state amplitude by measuring the largest absolute value of theta
    beyond the t=80 mark"""
    past_80 = dataframe[dataframe["t"] >= 80]
    return max(np.abs(past_80["theta_rk4"]))
```

```python
def analytical_steady_state_amplitude(g, L, F_D, Omega_D, q):
    return F_D / np.sqrt( (g/L - Omega_D**2)**2 + (q*Omega_D)**2  )
    
assa = np.vectorize(analytical_steady_state_amplitude)
```

```python
ls = np.linspace(0.1, 1.9, 100)

plt.subplots(figsize=(6.4*2, 4.8*2))
plt.suptitle("Amplitude of steady-state oscillation of a damped driven oscillator\ndriven at $\Omega_D$ with damping constant $q$", fontsize=16)


plt.subplot(221)
# q = 0.1

plt.plot(omegaDList[0], steady_state_amplitude(data[0.1][omegaDList[0]]), 'r.', label='Numerical')
for omega_d in omegaDList[1:]:
    plt.plot(omega_d, steady_state_amplitude(data[0.1][omega_d]), 'r.')

thetas = assa(10., 10., 1., ls, 0.1)
plt.plot(ls, thetas, label='Analytical')

plt.axvline(x=1, lw=0.5, color='g', label='Natural Frequency')

plt.ylim(0, 11)
plt.xlabel(r'$\Omega_D$')
plt.ylabel(r'$\theta_0$')
plt.title(r'$q=0.1$')
plt.legend()
    
    

plt.subplot(222)
# q = 0.2

plt.plot(omegaDList[0], steady_state_amplitude(data[0.2][omegaDList[0]]), 'r.', label='Numerical')
for omega_d in omegaDList[1:]:
    plt.plot(omega_d, steady_state_amplitude(data[0.2][omega_d]), 'r.')

thetas = assa(10., 10., 1., ls, 0.2)
plt.plot(ls, thetas, label='Analytical')

plt.axvline(x=1, lw=0.5, color='g', label='Natural Frequency')
    
plt.ylim(0, 11)
plt.xlabel(r'$\Omega_D$')
plt.ylabel(r'$\theta_0$')
plt.title(r'$q=0.2$')
plt.legend()
    

    
plt.subplot(223)
# q = 0.5

plt.plot(omegaDList[0], steady_state_amplitude(data[0.5][omegaDList[0]]), 'r.', label='Numerical')
for omega_d in omegaDList[1:]:
    plt.plot(omega_d, steady_state_amplitude(data[0.5][omega_d]), 'r.')
    
thetas = assa(10., 10., 1., ls, 0.5)
plt.plot(ls, thetas, label='Analytical')

plt.axvline(x=1, lw=0.5, color='g', label='Natural Frequency')
    
plt.ylim(0, 11)
plt.xlabel(r'$\Omega_D$')
plt.ylabel(r'$\theta_0$')
plt.title(r'$q=0.5$')
plt.legend()
    

    
plt.subplot(224)
# q = 1.0

plt.plot(omegaDList[0], steady_state_amplitude(data[1.0][omegaDList[0]]), 'r.', label='Numerical')
for omega_d in omegaDList[1:]:
    plt.plot(omega_d, steady_state_amplitude(data[1.0][omega_d]), 'r.')
    
thetas = assa(10., 10., 1., ls, 1.0)
plt.plot(ls, thetas, label='Analytical')

plt.axvline(x=1, lw=0.5, color='g', label='Natural Frequency')
    
plt.ylim(0, 11)
plt.xlabel(r'$\Omega_D$')
plt.ylabel(r'$\theta_0$')
plt.title(r'$q=1.0$')
plt.legend()

plt.show()
```

This plot shows the amplitude of the steady-state oscillation of a damped driven oscillator, $\theta_0$, over a range of driving frequencies $\Omega_D$ and at four different damping parameters $q$. The natural frequency of the pendulum, $\Omega = \sqrt{\frac{g}{l}}$, is set to 1.

Note that for small $q$, the plot of $\theta_0$ vs. $\Omega_D$ exhibits a sharp peak near the natural frequency. This means that when $\Omega_D \approx \Omega$, the steady state oscillation has a much larger amplitude that when the two frequencies don't match. This phenomenon is known as resonance. The analytical expectation for the resonance is also plotted (Eqn. 3.16 in Computational Physics), and we can see that our Order 4 Runge Kutta model obeys the analytical expectations well.

As $q$ gets larger, i.e. the pendulum gets more damped, the resonance peak flattens and then disappears. This observation is also supported by the analytical equation.
