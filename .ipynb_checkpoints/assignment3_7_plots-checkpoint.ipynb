```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/harmonic_oscillator.dat', index_col=0)
```

```python
df.head()
```

```python
plt.figure(figsize=(6.4 * 2, 4.8 * 2))

plt.subplot(221)
plt.plot(df.index, df["theta_rk2"], ls='none', marker='.', label = 'RK O(2)')
plt.plot(df.index, df["theta_rk4"], ls='none', marker='.', label = 'RK O(4)')
plt.plot(df.index, df["theta_euler"], ls='none', marker='.', label='euler')

plt.title("Simple harmonic oscillator")
plt.xlabel("$t$")
plt.ylabel(r"$\theta$")
plt.legend()


plt.subplot(222)
plt.plot(df.index, df["E_rk2"], ls='none', marker='.', label = 'RK O(2)')
plt.plot(df.index, df["E_rk4"], ls='none', marker='.', label = 'RK O(4)')
plt.plot(df.index, df["E_euler"], ls='none', marker='.', label='euler')

plt.title("Energy of simple harmonic oscillator")
plt.xlabel("$t$")
plt.ylabel(r"$E$")
plt.legend()


plt.subplot(223)
plt.plot(df.index, df["E_rk2"], ls='none', marker='.', label = 'RK O(2)')
plt.plot(df.index, df["E_rk4"], ls='none', marker='.', label = 'RK O(4)')

plt.title("Energy of simple harmonic oscillator")
plt.xlabel("$t$")
plt.ylabel(r"$E$")
plt.legend()
```

(Subplot 1) This plot shows the motion of a simple harmonic oscillator as solved by the Euler, Runge Kutta order 2, and Runge Kutta order 4 methods. Note the seemingly unphysical behavior of the Euler's method solution. This behavior is shown to be unphysical via an analysis of system energy in the next plot.

(Subplot 2) This plot shows the energy of the simple harmonic oscillator when solved by the Euler, Runge Kutta order
2, and Runge Kutta order 4 methods. Note that the energy of the Euler method solution appears to grow quadratically with time. This is clearly unphysical, since the energy of the simple harmonic oscillator should be conserved with respect to time! We conclude that Euler's method is unsuitable for solving the simple harmonic oscillator.

(Subplot 3) This plot shows the energy of the simple harmonic oscillator when solved by the Runge Kutta order
2 and Runge Kutta order 4 methods. Note that the energy of the Runge Kutta Order 2 method solution appears to grow linearly with time. Although not as severe as the Euler method solution, this behavior is clearly unphysical, since the energy of the simple harmonic oscillator should be conserved with respect to time! We conclude thatthe Runge Kutta order 2 method is unsuitable for solving the simple harmonic oscillator. In contrast, the energy using the Runge Kutta Order 4 method stays constant with respect to time, suggesting that this method is capable of suitably modeling the simple harmonic oscillator.
